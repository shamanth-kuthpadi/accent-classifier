### Loading Files

*Sample Loading of File*

```python
def load_audio_file(file_path, sr = 48000, fixed_duration = 2.5):   
    
    # Load the audio file with the specified sampling rate
    y, sr = librosa.load(file_path, sr=sr)
    
    # Calculate the target number of samples for the fixed duration
    target_length = int(fixed_duration * sr)
    
    # Adjust the length of the time series - looking into how the length is being fixed
    y = librosa.util.fix_length(y, size = target_length)

    #print(f"Audio Time-Series: {y.shape}")
    #print(f"Sampling Rate: {sr}")
    #print(f"Fixed Duration (samples): {target_length}")

    return y, sr
```

```python
sample_y, sample_sr = load_audio_file('ProcessedData/taru_reveal.wav')
```

### Extracting Features

```python
def extract_mfcc(y, sr, n_mfcc=13):
    
    """
    Extract Mel-frequency cepstral coefficients (MFCCs) from an audio time-series.

    Parameters:
        y (numpy.ndarray): Audio time-series.
        sr (int): Sampling rate of the audio.
        n_mfcc (int): Number of MFCCs to return. Default is 13.

    Returns:
        numpy.ndarray: MFCCs array of shape (n_mfcc, time_frames).
    """

    # Computing the MFCCs
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    # print(f"MFCCs Shape: {mfccs.shape}")
    return mfccs
```

```python
# Extracting MFCCs
sample_mfccs = extract_mfcc(sample_y, sample_sr)
print(sample_mfccs)
```

```python
def extract_delta_mfcc(mfccs):
    """
    Extract delta MFCCs (the first-order derivatives of the MFCCs).
    
    Parameters:
        mfccs (numpy.ndarray): The MFCCs array of shape (n_mfcc, time_frames).
    
    Returns:
        numpy.ndarray: The delta MFCCs array.
    """
    # Compute the delta (first derivative) of the MFCCs
    delta_mfccs = librosa.feature.delta(mfccs)
    # print(f"Delta MFCCs Shape: {delta_mfccs.shape}")
    return delta_mfccs
```

```python
# Extracting Delta MFCCs
sample_delta_mfccs = extract_delta_mfcc(sample_mfccs)
```

```python
def extract_spectral_centroid(y, sr):
    
    """
    Extracts Spectral Centroid from an audio signal.
    
    Parameters:
        y (numpy.ndarray): Audio time-series.
        sr (int): Sampling rate of the audio signal.
    
    Returns:
        numpy.ndarray: Spectral centroid of the signal.
    """
    spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
    # print(f"Spectral Centroid Shape: {spectral_centroid.shape}")
    return spectral_centroid
```

```python
# Extracting Spectral Centroid
sample_spectral_centroid = extract_spectral_centroid(sample_y, sample_sr)
```

```python
def extract_tonnetz(y, sr):
    """
    Extracts Tonnetz (Harmonic Features) from an audio signal.
    
    Parameters:
        y (numpy.ndarray): Audio time-series.
        sr (int): Sampling rate of the audio signal.
    
    Returns:
        numpy.ndarray: Tonnetz features.
    """
    tonnetz = librosa.feature.tonnetz(y=y, sr=sr)
    # print(f"Tonnetz Shape: {tonnetz.shape}")
    return tonnetz
```

```python
# Extracting Tonnetz Features
sample_tonnetz = extract_tonnetz(sample_y, sample_sr)
```

```python
def extract_features(y, sr):

    """
    Extracts various audio features from an audio signal.
    
    Parameters:
        y (numpy.ndarray): Audio time-series.
        sr (int): Sampling rate of the audio signal.
    
    Returns:
        dict: Extracted features.
    """
    
    # Extracting MFCCs
    mfccs = extract_mfcc(y, sr)
    
    # Extracting Delta MFCCs
    delta_mfccs = extract_delta_mfcc(mfccs)
    
    # Extracting Spectral Centroid
    spectral_centroid = extract_spectral_centroid(y, sr)
    
    # Extracting Tonnetz Features
    tonnetz = extract_tonnetz(y, sr)
    
    # Stack all features together
    features = np.vstack([mfccs, delta_mfccs, spectral_centroid, tonnetz])

    # normalize the features
    # NOTE : scope to normalize each feature type seperately using specific techniques for each type.
    features = StandardScaler().fit_transform(features)

    return features.T  # Transpose to have samples as rows
```

```python
# Extracting Features
sample_features = extract_features(sample_y, sample_sr)
print(sample_features.shape)
```

```python
root_directory = 'ProcessedData'
audio_files = glob.glob(os.path.join(root_directory, '**', '*.wav'), recursive=True)
```

```python
# creating a function to extract features and put them into X for feature set and y for label set
def extract_feature_set(audio_files):

    X_list = []
    y_list = []

    # audio_files = glob.glob(os.path.join(root_directory, '**', '*.wav'), recursive=True)

    for file_path in audio_files:

        # Load the audio file
        y, sr = load_audio_file(file_path)

        # Extract features
        features = extract_features(y, sr)

        # Append features to the list
        X_list.append(features)
        
        # Extract the label from the file path - everything after the last '/' and before the '.'
        file_name = file_path.split('/')[-1].split('.')[0]

        # NOTE : 1 for Taru and 0 for Shamanth
        if 'taru' in file_name:
            y_list.append(1)
        else:
            y_list.append(0)

    # stack the list of feature arrays into a 3d tensor of shape (n_samples, n_time_frames, n_features)
    X = np.dstack(X_list)

    # reshape X where the number of samples is the first dimension
    X = X.transpose((2, 0, 1))

    # Convert the label list to a numpy array
    y = np.array(y_list)

    # reshape y to have dimensions of (n_samples, 1)
    y = y.reshape(-1, 1)

    print(f"X Shape: {X.shape}")
    print(f"y Shape: {y.shape}")

    return X, y
```
